```python
all_news_final_fake = pd.read_csv('C:/Users/igrew/OneDrive/Desktop/CourseWork/datasets1/ddatasets/fake_news/all_news_final_fake.csv')
data_set_4_fake = pd.read_csv('C:/Users/igrew/OneDrive/Desktop/CourseWork/datasets1/ddatasets/fake_news/data_set_4_fake.csv')
news_data_fake = pd.read_csv('C:/Users/igrew/OneDrive/Desktop/CourseWork/datasets1/ddatasets/fake_news/news_data_fake.csv')
```

```python
all_news_final_fake = all_news_final_fake.rename(columns={'ukr_text': 'text'})
if 'id' in all_news_final_fake.columns:
    all_news_final_fake = all_news_final_fake.drop(columns=['id'])
all_news_final_fake['label'] = all_news_final_fake['label'].map({False: 1, True: 0})
all_news_final_fake = all_news_final_fake[['text', 'label']]

all_news_final_fake.to_csv('C:/Users/igrew/OneDrive/Desktop/CourseWork/datasets1/ddatasets/fake_news/all_news_final_fake.csv', index=False, encoding='utf-8')
all_news_final_fake.head()
```

```python
columns_to_drop = ['Unnamed: 0', 'Link', 'id']
for col in columns_to_drop:
    if col in data_set_4_fake.columns:
        data_set_4_fake = data_set_4_fake.drop(columns=[col])

data_set_4_fake = data_set_4_fake.rename(columns={'Text': 'text', 'Label': 'label'})

data_set_4_fake['label'] = data_set_4_fake['label'].map({False: 1, True: 0})

data_set_4_fake = data_set_4_fake[['text', 'label']]
data_set_4_fake.to_csv('C:/Users/igrew/OneDrive/Desktop/CourseWork/datasets1/ddatasets/fake_news/data_set_4_fake.csv', index=False, encoding='utf-8')

print(f"data_set_4_fake очищено: {len(data_set_4_fake)} рядків")
data_set_4_fake.head()
```

```python
columns_to_drop = ['Unnamed: 0', 'Link', 'id']
for col in columns_to_drop:
    if col in news_data_fake.columns:
        news_data_fake = news_data_fake.drop(columns=[col])

news_data_fake = news_data_fake.rename(columns={'Text': 'text', 'Label': 'label'})

news_data_fake['label'] = news_data_fake['label'].map({False: 1, True: 0})

news_data_fake = news_data_fake[['text', 'label']]
news_data_fake.to_csv('C:/Users/igrew/OneDrive/Desktop/CourseWork/datasets1/ddatasets/fake_news/news_data_fake.csv', index=False, encoding='utf-8')

print(f"news_data_fake очищено: {len(news_data_fake)} рядків")
news_data_fake.head()
```

```python
full_dataset = pd.read_csv('C:/Users/igrew/OneDrive/Desktop/CourseWork/Models/DETECTOR/full_dataset.csv')
print(f"Початковий full_dataset: {len(full_dataset)} рядків")

combined = pd.concat([full_dataset, all_news_final_fake, data_set_4_fake, news_data_fake], ignore_index=True)
print(f"Після об'єднання: {len(combined)} рядків")

combined = combined.drop_duplicates(subset=['text'], keep='first')
print(f"Після видалення дублікатів: {len(combined)} рядків")

combined.to_csv('C:/Users/igrew/OneDrive/Desktop/CourseWork/Models/DETECTOR/full_dataset.csv', index=False, encoding='utf-8')
print("full_dataset.csv оновлено")
combined.head()
```

```python
all_news_final = pd.read_csv('C:/Users/igrew/OneDrive/Desktop/CourseWork/datasets1/ddatasets/all_news_final.csv')
print(f"Завантажено all_news_final: {len(all_news_final)} рядків")

true_news = all_news_final[all_news_final['label'] == True].copy()
print(f"Правдивих новин (True): {len(true_news)}")

true_news = true_news.rename(columns={'ukr_text': 'text'})
true_news['label'] = 0
true_news = true_news[['text', 'label']]

current_full = pd.read_csv('C:/Users/igrew/OneDrive/Desktop/CourseWork/Models/DETECTOR/full_dataset.csv')
print(f"\nПоточний full_dataset: {len(current_full)} рядків")
print(f"Fake: {(current_full['label']==1).sum()}, True: {(current_full['label']==0).sum()}")

existing_texts = set(current_full['text'].values)
true_news_unique = true_news[~true_news['text'].isin(existing_texts)].copy()
print(f"\nУнікальних правдивих новин (яких немає в датасеті): {len(true_news_unique)}")

needed_count = 7059 - 982
if len(true_news_unique) < needed_count:
    print(f"Доступно тільки {len(true_news_unique)} унікальних новин, а потрібно {needed_count}")
    true_news_sample = true_news_unique
else:
    true_news_sample = true_news_unique.head(needed_count)

print(f"Взято для додавання: {len(true_news_sample)} унікальних рядків")

extended = pd.concat([current_full, true_news_sample], ignore_index=True)
print(f"\nПісля додавання: {len(extended)} рядків")
print(f"Fake: {(extended['label']==1).sum()}, True: {(extended['label']==0).sum()}")

extended.to_csv('C:/Users/igrew/OneDrive/Desktop/CourseWork/Models/DETECTOR/full_dataset.csv', index=False, encoding='utf-8')
print("\nfull_dataset.csv оновлено з правдивими новинами")
extended.head()
```
